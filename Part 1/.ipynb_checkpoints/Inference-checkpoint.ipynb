```python
# Import required libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

def load_test_data():
    """
    Load test data for inference.
    
    Returns:
        tuple: (X_test, y_test) containing test images and their labels
    """
    # Load a portion of data for testing
    test_data = np.load('../Test_data/data9 (1).npy')
    test_labels = np.load('../Test_data/label9.npy')
    
    return test_data, test_labels

def preprocess_image(image):
    """
    Preprocess a single image or batch of images.
    
    Args:
        image (numpy.ndarray): Input image(s)
        
    Returns:
        numpy.ndarray: Preprocessed image(s)
    """
    if len(image.shape) == 2:
        image = image.reshape(1, image.shape[0], image.shape[1], 1)
    else:
        image = image.reshape(-1, image.shape[1], image.shape[2], 1)
    return image / 255.0

def calculate_accuracy(y_true, y_pred, tolerance=0.5):
    """
    Calculate prediction accuracy with different tolerance levels.
    
    Args:
        y_true (numpy.ndarray): True labels
        y_pred (numpy.ndarray): Predicted labels
        tolerance (float): Tolerance for considering prediction as correct
        
    Returns:
        tuple: (exact_accuracy, tolerance_accuracy)
    """
    exact_accuracy = np.mean(np.round(y_pred) == y_true)
    tolerance_accuracy = np.mean(np.abs(y_true - y_pred) <= tolerance)
    
    return exact_accuracy, tolerance_accuracy

# Load the saved model
try:
    model = load_model('digit_sum_model.keras')
except:
    print("Error: Make sure to run training.ipynb first to generate the model!")
    raise

# Load and preprocess test data
X_test, y_test = load_test_data()
X_test = preprocess_image(X_test)

# Make predictions
y_pred = model.predict(X_test)

# Calculate accuracies
exact_acc, tolerance_acc = calculate_accuracy(y_test, y_pred.flatten())

print("\nModel Evaluation Results:")
print(f"Exact Match Accuracy: {exact_acc:.4f}")
print(f"Accuracy (within ±0.5): {tolerance_acc:.4f}")

# Evaluate model with compiled metrics
test_loss, test_mae = model.evaluate(X_test, y_test, verbose=0)
print(f"\nTest Loss (MSE): {test_loss:.4f}")
print(f"Test MAE: {test_mae:.4f}")

# Visualize some predictions
def plot_predictions(X_test, y_test, y_pred, num_samples=5):
    """
    Plot sample images with their true and predicted values.
    
    Args:
        X_test (numpy.ndarray): Test images
        y_test (numpy.ndarray): True labels
        y_pred (numpy.ndarray): Predicted labels
        num_samples (int): Number of samples to plot
    """
    plt.figure(figsize=(15, 3))
    for i in range(num_samples):
        plt.subplot(1, num_samples, i+1)
        plt.imshow(X_test[i].reshape(X_test[i].shape[0], X_test[i].shape[1]), cmap='viridis')
        plt.title(f'True: {y_test[i]}\nPred: {y_pred[i][0]:.1f}')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Plot sample predictions
plot_predictions(X_test, y_test, y_pred)

# Print confusion matrix-like analysis
def analyze_predictions(y_true, y_pred, tolerance=0.5):
    """
    Analyze predictions in detail.
    
    Args:
        y_true (numpy.ndarray): True labels
        y_pred (numpy.ndarray): Predicted labels
        tolerance (float): Tolerance for considering prediction as correct
    """
    exact_matches = np.sum(np.round(y_pred) == y_true)
    within_tolerance = np.sum(np.abs(y_true - y_pred) <= tolerance)
    total = len(y_true)
    
    print("\nDetailed Prediction Analysis:")
    print(f"Total Predictions: {total}")
    print(f"Exact Matches: {exact_matches} ({exact_matches/total*100:.2f}%)")
    print(f"Within ±{tolerance} tolerance: {within_tolerance} ({within_tolerance/total*100:.2f}%)")
    print(f"Average Absolute Error: {np.mean(np.abs(y_true - y_pred)):.3f}")

# Analyze predictions
analyze_predictions(y_test, y_pred.flatten())
```
